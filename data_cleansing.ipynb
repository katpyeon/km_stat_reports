# 타겟 데이터 정리
- 서울 데이터 필터링: 시군구별 열에서 서울특별시에 해당하는 데이터만 필터링.
- 긴(long) 형태로 변환: 연도별 데이터를 "Year", "출생아수(Births)", "합계출산율(Total_Fertility_Rate)"로 재구성.
- 데이터 타입 변환: 연도(Year), 출생아수(Births), 합계출산율(Total_Fertility_Rate)을 숫자형으로 변환.
- 컬럼명 변경: 한글 컬럼명을 영어로 변경(출생아수 → Births, 합계출산율 → Total_Fertility_Rate).

````python
# Import required libraries
import pandas as pd

# Specify file paths
data_path = './data/타겟_시군구_출생아수_합계출산율.csv'  # Input file path
output_path = './out/birth_rate.csv'  # Output file path

# Read the CSV file
birth_data = pd.read_csv(data_path)

# Filter for Seoul only
seoul_data = birth_data[birth_data['시군구별'] == '서울특별시']

# Reshape to long format
long_format = seoul_data.melt(id_vars=['시군구별'], var_name='Year_Variable', value_name='Value')

# Extract Year and Variable
long_format['year'] = long_format['Year_Variable'].str.extract(r'(\d{4})')  # Extract year
long_format['Variable'] = long_format['Year_Variable'].apply(lambda x: 'birth_count' if '.' not in x else 'birth_rate')

# Pivot data to create separate columns for Births and Total Fertility Rate
seoul_cleaned = long_format.pivot(index='year', columns='Variable', values='Value').reset_index()

# Convert columns to numeric where appropriate
seoul_cleaned['birth_count'] = pd.to_numeric(seoul_cleaned['birth_count'], errors='coerce')
seoul_cleaned['birth_rate'] = pd.to_numeric(seoul_cleaned['birth_rate'], errors='coerce')

# Save the cleaned data to a new CSV file
seoul_cleaned.to_csv(output_path, index=False)

# Print a preview of the cleaned data
display(seoul_cleaned.head())
display(seoul_cleaned.info())   
display(seoul_cleaned.describe())

````

# 고용임금 데이터 정리
- 불필요 행 제거: 첫 번째 행(설명용 행)을 삭제.
- 긴(long) 형태로 변환: 연도별 데이터를 "Year", "Monthly_Wage", "Gender", "Employment_Type"로 재구성.
- 데이터 타입 변환: 연도(Year)와 월임금(Monthly_Wage)을 숫자형으로 변환.
- 컬럼명 변경:
    - 고용형태 → Employment_Type
    - 성별 → Gender
- Gender 값 변경:
    - 전체 → A
    - 남 → M
    - 여 → F
- 필터링: Employment_Type가 전체근로자인 데이터에서 Gender 값이 A, M, F인 데이터만 추출.

````python
# Full script to clean and transform the employment and wage data

# File paths
input_path_wage = './data/경제_성별_임금_및_근로시간.csv'
output_path_wage = './out/employment_wage_cleaned.csv'

# Load the data
employment_wage_data = pd.read_csv(input_path_wage)

# Drop the first row (header info) and reset index
employment_wage_data_cleaned = employment_wage_data.drop(0).reset_index(drop=True)

# Reshape the data into long format
long_format = employment_wage_data_cleaned.melt(
    id_vars=['고용형태', '성별'], 
    var_name='year', 
    value_name='monthly_wage'
)

# Convert the year and wage to appropriate types
long_format['year'] = pd.to_numeric(long_format['year'], errors='coerce')
long_format['monthly_wage'] = pd.to_numeric(long_format['monthly_wage'], errors='coerce')

# Rename columns to English
long_format.rename(columns={
    '고용형태': 'employment_type',
    '성별': 'gender'
}, inplace=True)

# Modify gender values
gender_mapping = {'전체': 'A', '남': 'M', '여': 'F'}
long_format['gender'] = long_format['gender'].map(gender_mapping)

# Filter for male and female data only
filtered_data_mf = long_format[
    long_format['gender'].isin(['M', 'F'])
]

# Drop 'employment_type' column
filtered_data_mf = filtered_data_mf.drop(columns=['employment_type'])

# Reorder columns to 'year', 'gender', 'monthly_wage'
filtered_data_mf = filtered_data_mf[['year', 'gender', 'monthly_wage']]

# Save the cleaned data with only male and female genders included
filtered_data_mf.to_csv(output_path_wage, index=False)

# Display a preview of the cleaned data
display(filtered_data_mf.head())
display(filtered_data_mf.info())   
display(filtered_data_mf.describe())
````

# 주택 매매가격 데이터 정리
- 서울 데이터 필터링: 지역별(1) 열에서 "서울" 데이터만 추출.
- 긴(long) 형태 변환: 연도와 월을 포함한 데이터를 Year_Month에서 Year로 분리.
- 연 단위로 통합: 월별 데이터를 연 평균으로 계산하여 Year 기준으로 통합.
- 컬럼명 변경: Price_Index → Average_Housing_Price_Index.

````python
# File paths
input_path_housing = './data/경제_유형별_매매가격지수.csv'
output_path_housing = './out/housing_price_index_cleaned.csv'

# Load the data
housing_price_data = pd.read_csv(input_path_housing)

# Filter for '서울' region only
seoul_data = housing_price_data[housing_price_data['지역별(1)'] == '서울']

# Reshape data from wide to long format
long_format = seoul_data.melt(
    id_vars=['주택유형별(1)', '지역별(1)'], 
    var_name='year_month', 
    value_name='price_index'
)

# Extract year from Year_Month and convert Price_Index to numeric
long_format['year'] = long_format['year_month'].str.extract(r'^(\d{4})').astype(int)
long_format['price_index'] = pd.to_numeric(long_format['price_index'], errors='coerce')

# Aggregate by year using mean
yearly_data = long_format.groupby(['year'], as_index=False)['price_index'].mean().round(2)

# Rename columns for clarity
yearly_data.rename(columns={'price_index': 'housing_price_index'}, inplace=True)

# Save the cleaned data
yearly_data.to_csv(output_path_housing, index=False)

# Display a preview of the cleaned data
display(yearly_data.head())
display(yearly_data.info())   
display(yearly_data.describe())
````

# 주택 전세가격 데이터 정리

- 서울 데이터 필터링: 지역별(1) 열에서 "서울" 데이터를 필터링.
- 긴(long) 형태로 변환: 연도와 월별 데이터를 Year_Month로 변환 후, 연도(Year)만 추출.
- 연 단위로 통합: 월별 데이터를 연 평균으로 계산하여 Year 기준으로 통합.
- 컬럼명 변경: Price_Index를 Average_Rental_Price_Index로 변경하여 매매가격지수와 구분.


````python
# Full script to process and aggregate rental price index data for logistic regression

# File paths
input_path_rental = './data/경제_유형별_전세가격지수.csv'
output_path_rental = './out/rental_price_index_cleaned.csv'

# Load the data
rental_price_data = pd.read_csv(input_path_rental)

# Filter for '서울' region only
seoul_rental_data = rental_price_data[rental_price_data['지역별(1)'] == '서울']

# Reshape data from wide to long format
long_format_rental = seoul_rental_data.melt(
    id_vars=['주택유형별(1)', '지역별(1)'], 
    var_name='year_month', 
    value_name='price_index'
)

# Extract year from Year_Month and convert Price_Index to numeric
long_format_rental['year'] = long_format_rental['year_month'].str.extract(r'^(\d{4})').astype(int)
long_format_rental['price_index'] = pd.to_numeric(long_format_rental['price_index'], errors='coerce')

# Aggregate by year using mean
yearly_rental_data = long_format_rental.groupby(['year'], as_index=False)['price_index'].mean().round(2)

# Rename columns for clarity
yearly_rental_data.rename(columns={'price_index': 'rental_price_index'}, inplace=True)

# Save the cleaned data
yearly_rental_data.to_csv(output_path_rental, index=False)

# Display a preview of the cleaned data
display(yearly_rental_data.head())
display(yearly_rental_data.info())   
display(yearly_rental_data.describe())
````

# 성별 고용률 데이터 정리
- CSV 파일을 로드하고, 두 번째 행(인구_고용율) 설명 행을 제거.
- 월별 데이터 제거:
- 연도별 데이터 쌍으로 처리:

````python
# File path
file_path = './data/경제_성별_고용률.csv'
output_path = './out/employment_population_cleaned.csv'

# Step 1: Load data and remove explanation row
data = pd.read_csv(file_path, header=None)
data = data.drop(index=1)  # Remove the explanation row
data.columns = data.iloc[0]  # Set the first row as column names
data = data[1:]  # Remove the header row

# Step 2: Filter relevant rows
data = data[data['년도'].isin(['남자', '여자'])]  # Keep only '남자' and '여자'

# Step 3: Remove monthly data
data = data.loc[:, ~data.columns.str.contains(r'\.')]  # Drop columns with '.'

# Step 4: Process yearly data
processed_data = []
columns = data.columns[1:]  # Exclude '년도'

for i in range(0, len(columns), 2):  # Process in pairs
    year = int(columns[i])  # Extract year from the column name
    population_col = columns[i]  # Left column is Population
    rate_col = columns[i + 1]  # Right column is Rate

    for idx, row in data.iterrows():
        # Correct value extraction using .at
        population_value = row.at[population_col]
        rate_value = row.at[rate_col]
        
        # Debug: Verify extracted values
        try:            
            population_value = float(population_value[0])
            rate_value = float(rate_value[1])
        except ValueError:
            print(f"Skipping invalid data: Population={population_value}, Rate={rate_value}")
            continue
        
        processed_data.append({
            'year': year,
            'gender': 'M' if row['년도'] == '남자' else 'F',
            'employment_population': population_value,
            'employment_rate': rate_value
        })

# Step 5: Convert to DataFrame
result_df = pd.DataFrame(processed_data)

# Step 6: Save the cleaned data
result_df.to_csv(output_path, index=False)

# Step 7: Display the cleaned data

display(result_df.head())
display(result_df.info())   
display(result_df.describe())
````

# 초혼 연령 데이터 정리 

연도 및 부부구분 처리:
- 열 이름에서 연도(2000, 2001, ...)와 부부구분(남편, 아내)을 분리.
- 남편 → H, 아내 → W로 변환.

값 추출 및 데이터 정리:
- 남편과 아내 데이터를 한 쌍으로 처리:
- 연도(Year)별로 Spouse_Type(H/W)와 First_Marriage_Age로 정리.
- First_Marriage_Age를 소수점 2자리로 제한.


````python
# File path
file_path = './data/사회_시도별_평균초혼연령.csv'
output_path = './out/first_marriage_age.csv'

# Step 1: Load and clean data
data = pd.read_csv(file_path, header=None)

# Step 2: Remove explanation row (second row)
data.columns = data.iloc[0]  # Set the first row as column names
data = data[2:]  # Remove the first two rows

# Step 3: Filter for 서울특별시
data = data[data['년도'] == '서울특별시']

# Step 4: Reshape data
# Extract years and spouse types from column names
columns = data.columns[1:]  # Exclude '시도별'
years = [int(col.split()[0]) for col in columns[::2]]  # Extract unique years
husband_columns = columns[::2]  # Odd-indexed columns (남편)
wife_columns = columns[1::2]  # Even-indexed columns (아내)

# Create a DataFrame for the final output
processed_data = []

for year, h_col, w_col in zip(years, husband_columns, wife_columns):
    # Extract single values using .at or specific index    
    husband_age = float(data.at[data.index[0], h_col][0])  # Access first row and husband column
    wife_age = float(data.at[data.index[0], w_col][1])     # Access first row and wife column
    
    processed_data.append({'year': year, 'gender': 'M', 'first_marriage_age': husband_age})
    processed_data.append({'year': year, 'gender': 'F', 'first_marriage_age': wife_age})

# Convert to DataFrame
processed_df = pd.DataFrame(processed_data)

# Step 5: Round First_Marriage_Age to 2 decimal places
processed_df['first_marriage_age'] = processed_df['first_marriage_age'].round(2)

# Step 6: Save the processed data
processed_df.to_csv(output_path, index=False)

# Display the processed data
display(processed_df.head())
display(processed_df.info())   
display(processed_df.describe())
````

# 교육정도별 혼인연령 데이터 정리 

데이터 변환:
- melt를 사용하여 데이터를 Year_Spouse, Marriage_Count 형식으로 변환.
- Year_Spouse에서 연도(Year)와 부부구분(남편/아내)을 추출.
- cumcount()를 사용하여 Spouse_Type을 결정:
- 0번째 값: H(남편).
- 1번째 값: W(아내).

컬럼명 변경 및 정리:
- 연령별 → Age_Group, 교육정도별 → Education_Level, Marriage_Count → Marriage_Count.
- 최종 데이터: Year, Spouse_Type, Age_Group, Education_Level, Marriage_Count.

````python
# Corrected File path
file_path = './data/사회_연령_교육정도별_혼인.csv'
output_path = './out/age_education_marriage_paired.csv'

# Step 1: Load data
data = pd.read_csv(file_path, header=None)

# Step 2: Remove explanation row (second row)
data.columns = data.iloc[0]  # Set the first row as column names
data = data[2:]  # Remove the first two rows

# Step 3: Remove rows where '교육정도별' is '총계'
data = data[data['교육정도별'] != '총계']

# Step 4: Reshape data to long format
data_long = data.melt(id_vars=['연령별', '교육정도별'], var_name='year_spouse', value_name='marriage_count')

# Replace '-' with 0 and convert to numeric
data_long['marriage_count'] = data_long['marriage_count'].replace('-', 0).astype(float)

# Step 5: Extract year
data_long['year'] = data_long['year_spouse'].str.extract(r'(\d{4})').astype(int)

# Determine gender based on pairing order
data_long['gender'] = data_long.groupby(['year', '연령별', '교육정도별']).cumcount().map({0: 'M', 1: 'F'})

# Step 6: Rename columns for clarity (to snake_case)
data_long.rename(columns={
    '연령별': 'age_group',
    '교육정도별': 'education_level'
}, inplace=True)

# Step 7: Keep relevant columns and save data
processed_data = data_long[['year', 'gender', 'age_group', 'education_level', 'marriage_count']]
processed_data.to_csv(output_path, index=False)

# Display the processed data
print(processed_data["gender"].value_counts())
print(processed_data[processed_data["year"] == 2000])

display(processed_data)
display(processed_data.info())   
display(processed_data.describe())
````

	9.	first_marriage_age (평균 초혼 연령): 서울특별시의 연도별 성별 평균 초혼 연령.
	10.	urbanized_area_rate (시가화율): 서울특별시의 연도별 도시화 정도.
	11.	edu_index (학력 지수): 서울특별시의 연도별 성별 학력 지수.

3. 분석 데이터 요약

	1.	결측값 처리 완료:
	•	병합된 모든 데이터는 결측값 없이 완전한 상태.
	2.	분석 대상:
	•	서울특별시의 출산율과 경제적, 사회적 요인의 관계를 분석하기 위한 최적화된 데이터.
	3.	저장 경로:
	•	병합된 데이터는 ./cleaned/final_data.csv로 저장되어 재사용 가능.

추가 활용 방안

	•	employment_rate, monthly_wage, first_marriage_age 등 독립 변수를 활용한 예측 모델 설계 가능.

````python
import pandas as pd

# 파일 경로 설정
file_paths = {
    "age_education_marriage": "./out/age_education_marriage_paired.csv",
    "birth_rate": "./out/birth_rate.csv",
    "employment_population": "./out/employment_population_cleaned.csv",
    "employment_wage": "./out/employment_wage_cleaned.csv",
    "first_marriage_age": "./out/first_marriage_age.csv",
    "housing_price_index": "./out/housing_price_index_cleaned.csv",
    "rental_price_index": "./out/rental_price_index_cleaned.csv",
    "urbanized_area_rate": "./out/seoul_urbanized_area_rate.csv"
}

# 데이터 로드
dataframes = {name: pd.read_csv(path) for name, path in file_paths.items()}

# Step 1: 학력 데이터 보정
# 학력 매핑 설정
education_mapping = {
    '초등학교': 1,
    '중학교': 2,
    '고등학교': 3,
    '대학(교)': 4,
    '대학원 이상': 5
}

# 학력 데이터 로드 및 매핑 적용
education_data = dataframes["age_education_marriage"]
education_data['education_level_mapped'] = education_data['education_level'].map(education_mapping)

# Pivot table to organize data for interpolation
education_pivot = education_data.pivot_table(
    index=['year', 'gender'], 
    columns='education_level', 
    values='marriage_count', 
    aggfunc='sum'
)

# Fill missing values with 0 for interpolation preparation
education_pivot = education_pivot.fillna(0)

# Apply interpolation only to the "대학(교)" and "대학원 이상" columns
for level in ['대학(교)', '대학원 이상']:
    education_pivot[level] = education_pivot[level].replace(0, pd.NA)  # Replace 0 with NA
    education_pivot[level] = education_pivot[level].interpolate(method='linear', axis=0)

# Reset index for further processing
education_pivot = education_pivot.reset_index()

# Melt back to long format
education_corrected = education_pivot.melt(
    id_vars=['year', 'gender'], 
    var_name='education_level', 
    value_name='marriage_count'
)

# Map education levels back to ordinal values
education_corrected['education_level_mapped'] = education_corrected['education_level'].map(education_mapping)

# Step 2: 학력 지수 계산 (보정된 데이터 활용)
weighted_edu_index = (
    education_corrected
    .groupby(['year', 'gender'])
    .apply(lambda group: (group['education_level_mapped'] * group['marriage_count']).sum() / group['marriage_count'].sum())
    .reset_index(name='edu_index')
)

# Step 3: 병합 시작 (출생률 데이터 기준)
merged_data = dataframes["birth_rate"]

# 'employment_population' 병합
merged_data = pd.merge(merged_data, dataframes["employment_population"], on=["year"], how="outer")

# 'employment_wage' 병합
merged_data = pd.merge(merged_data, dataframes["employment_wage"], on=["year", "gender"], how="outer")

# 'first_marriage_age' 병합
merged_data = pd.merge(merged_data, dataframes["first_marriage_age"], on=["year", "gender"], how="outer")

# 'housing_price_index' 병합 (성별 없음)
merged_data = pd.merge(merged_data, dataframes["housing_price_index"], on=["year"], how="outer")

# 'rental_price_index' 병합 (성별 없음)
merged_data = pd.merge(merged_data, dataframes["rental_price_index"], on=["year"], how="outer")

# 'urbanized_area_rate' 병합 (성별 없음)
merged_data = pd.merge(merged_data, dataframes["urbanized_area_rate"], on=["year"], how="outer")

# 'edu_index' 병합
merged_data = pd.merge(merged_data, weighted_edu_index, on=["year", "gender"], how="left")

# 컬럼 이름을 소문자 및 스네이크 케이스로 변환
merged_data.columns = merged_data.columns.str.lower().str.replace(" ", "_")

# 결측값 처리: 수치 데이터는 선형 보간법(interpolation) 적용
merged_data.sort_values(by="year", inplace=True)
merged_data.interpolate(method="linear", inplace=True)

# 결측값 제거 후 데이터 저장
output_path = "./cleaned/final_data.csv"
merged_data.dropna(inplace=True)
merged_data.to_csv(output_path, index=False)

# 데이터 요약 및 확인
print("Merged data saved to:", output_path)
print(merged_data.info())
print(merged_data.head())
````
